## Analyzing Brand Attachment from Sample Orders
We want to determine whether customers who received a sample subsequently purchase products from the same brand. The following steps do this:

* Filter the sample orders.
* For each sample order, check if the same customer later made a non-sample purchase from that brand.
* Calculate the conversion rate (i.e., the percentage of sample orders that resulted in a brand purchase).

```python
# Filter the orders that are sample orders
sample_df = df[df['is_sample']].copy()

def check_brand_conversion(row, orders_df):
    """
    Given a sample order row, check if the customer later purchased a product
    from the same vendor (acting as the brand) excluding any further sample orders.
    """
    customer = row['customer_id']
    vendor = row['vendor']
    sample_date = row['Order_date']
    
    # Find later orders by the same customer
    later_orders = orders_df[(orders_df['customer_id'] == customer) & (orders_df['Order_date'] > sample_date)]
    # Consider only non-sample orders for conversion
    conversion = later_orders[(later_orders['vendor'] == vendor) & (~later_orders['is_sample'])]
    
    return 1 if not conversion.empty else 0

# Apply the function to each sample order
sample_df['brand_conversion'] = sample_df.apply(lambda row: check_brand_conversion(row, df), axis=1)

# Calculate the attachment rate for the brand (percentage of sample orders converting to vendor purchase)
brand_attachment_rate = sample_df['brand_conversion'].mean()
print("Brand Attachment Rate (conversion rate):", brand_attachment_rate)
```

```python
def check_product_conversion(row, orders_df):
    """
    Check if the customer purchased the same product (non-sample) after receiving a sample.
    """
    customer = row['customer_id']
    product = row['productName']
    sample_date = row['Order_date']
    
    # Find later orders by the same customer
    later_orders = orders_df[(orders_df['customer_id'] == customer) & (orders_df['Order_date'] > sample_date)]
    # Check if any of the later orders (non-sample) is for the same product
    conversion = later_orders[(later_orders['productName'] == product) & (~later_orders['is_sample'])]
    
    return 1 if not conversion.empty else 0

# Apply the function to each sample order
sample_df['product_conversion'] = sample_df.apply(lambda row: check_product_conversion(row, df), axis=1)

# Calculate the attachment rate for the product
product_attachment_rate = sample_df['product_conversion'].mean()
print("Product Attachment Rate (conversion rate):", product_attachment_rate)
```

```python
def conversion_time(row, orders_df):
    """
    Calculate the time (in days) between the sample order and the first subsequent purchase
    of the same product (excluding sample orders).
    """
    customer = row['customer_id']
    product = row['productName']
    sample_date = row['Order_date']
    
    # Get later orders by the customer
    later_orders = orders_df[(orders_df['customer_id'] == customer) & (orders_df['Order_date'] > sample_date)]
    # Filter for non-sample orders of the same product
    conversion_orders = later_orders[(later_orders['productName'] == product) & (~later_orders['is_sample'])]
    
    if conversion_orders.empty:
        return np.nan  # no conversion found
    # Get the earliest conversion order
    first_purchase = conversion_orders.sort_values(by='Order_date').iloc[0]
    return (first_purchase['Order_date'] - sample_date).days

# Apply the conversion time function to each sample order
sample_df['conversion_time_days'] = sample_df.apply(lambda row: conversion_time(row, df), axis=1)

# Calculate the average conversion time (in days) excluding cases with no conversion
avg_conversion_time = sample_df['conversion_time_days'].dropna().mean()
print("Average Conversion Time (days):", avg_conversion_time)
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(sample_df['conversion_time_days'].dropna(), bins=25, kde=True)
plt.xlabel("Conversion Time (Days)")
plt.title("Distribution of Conversion Time from Sample to Purchase")
plt.show()
```

```python
# Create a summary DataFrame for conversion rates
conversion_summary = pd.DataFrame({
    'Metric': ['Brand Attachment Rate', 'Product Attachment Rate'],
    'Conversion Rate': [brand_attachment_rate, product_attachment_rate]
})

plt.figure(figsize=(6, 4))
sns.barplot(x='Metric', y='Conversion Rate', data=conversion_summary)
plt.ylim(0, 1)
plt.title("Conversion Rates for Samples")
plt.show()
```
